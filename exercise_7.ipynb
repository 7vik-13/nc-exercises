```python
import matplotlib.pyplot as plt
import math
```

```python
# Question 5 and 6
class Point():
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def plot_point(self):
        plt.plot(self.x,self.y,"o")

        plt.grid()
        plt.show()
    
    def calculate_dist(self,x,y):
        print(math.sqrt((self.x-x)**2 + (self.y-y)**2))
```

```python
k = Point(3,4)
k.plot_point()
```

```python
k.calculate_dist(7,11)
```
