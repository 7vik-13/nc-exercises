```python
import numpy as np
import matplotlib.pyplot as plt 
import math
import random
```

```python
# Question 8
def my_fun_plotter(f, x):
    plt.plot(x, f(x))
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.show()
```

```python
my_fun_plotter(lambda x: np.sqrt(x) + np.exp(np.sin(x)), np.linspace(0, 2*np.pi, 100))
```

```python
# Question 9
def my_poly_plotter(n,x):
    for i in range(1, n+1):
        plt.plot(x, x**i)
    plt.title(f'Polynomials up to degree {n}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
my_poly_plotter(5, np.linspace(-1, 1, 200))
```

```python
# Question 10
def my_sierpinski(n):
    p1 = [0,0]
    p2 = [0.5, math.sqrt(3)/2]
    p3 = [1,0]
    p = [p3, p1, p2]
    p_i = [0,0]
    for i in p:
        plt.plot(i[0], i[1], 'b.')
    for _ in range(n):
        r = random.random()
        k = math.floor(3*r)
        plt.plot((p_i[0]+p[k][0])/2, (p_i[1]+p[k][1])/2, 'b.')
        p_i = [(p_i[0]+p[k][0])/2, (p_i[1]+p[k][1])/2]
    plt.axis("off")
    plt.axis("equal")
    plt.show()
```

```python
my_sierpinski(500)
```

```python
my_sierpinski(3000)
```

```python
# Question 11
def my_fern(n):
    p = [0, 0]
    plt.plot(0, 0, 'g.')
    for _ in range(n):
        r = random.random()
        if r <= 0.01:
            plt.plot(0, 0.16*p[1], 'g.')
            p = [0, 0.16*p[1]]
        elif r <= 0.08:
            plt.plot(0.2*p[0]-0.26*p[1], 0.23*p[0]+0.22*p[1]+1.6, 'g.')
            p = [0.2*p[0]-0.26*p[1], 0.23*p[0]+0.22*p[1]+1.6]
        elif r <= 0.15:
            plt.plot(-0.15*p[0]+0.28*p[1], 0.26*p[0]+0.24*p[1]+0.44, 'g.')
            p = [-0.15*p[0]+0.28*p[1], 0.26*p[0]+0.24*p[1]+0.44]
        else:
            plt.plot(0.85*p[0]+0.04*p[1], -0.04*p[0]+0.85*p[1]+1.6, 'g.')
            p = [0.85*p[0]+0.04*p[1], -0.04*p[0]+0.85*p[1]+1.6]
    plt.axis("off")
    plt.axis("equal")
    plt.show()
```

```python
my_fern(100)
```

```python
my_fern(5000)
```
